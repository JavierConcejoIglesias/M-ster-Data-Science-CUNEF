```python
import pandas as pd
import spacy
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from spacy import displacy
from IPython.display import display_html
from itertools import chain,cycle

nlp= spacy.load('es_core_news_lg')

def display_side_by_side(*args,titles=cycle([''])):
    html_str=''
    for df,title in zip(args, chain(titles,cycle(['</br>'])) ):
        html_str+='<th style="text-align:center"><td style="vertical-align:top">'
        html_str+=f'<h2 style="text-align: center;">{title}</h2>'
        html_str+=df.to_html().replace('table','table style="display:inline"')
        html_str+='</td></th>'
    display_html(html_str,raw=True)

import warnings
warnings.filterwarnings("ignore")
```

### Explicación

En primer lugar cargamos los dos ficheros de texto. He decidido convertir todos los documentos del fichero 'hotel' a minúsculas puesto que se pide en una actividad posterior pero considero que lo más lógico es realizar esta tarea en la carga y tratamiento de datos inicial.

## 2.  Lea el primer fichero, hotel.csv en un dataframe y realice las labores necesarias para obtener las matrices BOW y Tf-idf

```python
hotel=pd.DataFrame(hotel)
lista_hotel=list(hotel['text'])
```

#### BOW

```python
BOW_vectorizer = CountVectorizer()
BOW_array = BOW_vectorizer.fit_transform(lista_hotel).toarray()
vocab_BOW = BOW_vectorizer.get_feature_names()
BOW=pd.DataFrame(BOW_array, columns=vocab_BOW)
BOW
```

#### Tf-idf

```python
TFIDF_vectorizer = TfidfVectorizer()
TFIDF_array=TFIDF_vectorizer.fit_transform(lista_hotel).toarray()
vocab_TFIDF = TFIDF_vectorizer.get_feature_names()
TFIDF=pd.DataFrame(TFIDF_array, columns=vocab_TFIDF)
TFIDF
```

### Explicación

Para poder realizar las matrices BOW y Tfidf he tenido que transformar el texto del hotel en una lista la cual contiene todos los comentarios de las reseñas. Cada elemento de esta lista representa un comentario o documento del fichero "hotel". 

Las filas de la matriz BOW representan cada uno de los documentos y las columnas cada uno de los términos. El valor de las celdas representará el número de veces que un término aparece en ese documento. Por ejemplo, el término "10" no aparece en el documento 0 (primer comentario) pero si en el documento 1 (segundo comentario) una vez. Importante destacar que el BOW no tiene en cuenta el orden o secuencia de las palabras.

La matriz Tfidf sigue la misma estrucutura que la de BOW. Sin embargo, los valores de las celdas de esta matriz no indican el número de veces que aparece el término en el documento ya que este tipo de modelo de lenguaje soluciona el problema que tiene el BOW de dar mas importancia a aquellos términos que son más repetidos con respecto a otros términos que no son tan repetidos pero si pueden ser relevantes. Por eso, el modelo Tfidf realiza una normalización. Los valores de las celdas son los pesos de cada término, los cuales indican la frecuencia de ese término en ese documento con respecto a la frecuencia del mismo término en el corpus completo. 

## 3.  Para estas matrices, utilice únicamente los 20 tokens más frecuentes y emplee la conversión de todos los términos a minúsculas

```python
total_BOW = pd.DataFrame(BOW.sum(axis=0),columns=['Apariciones'])
total_BOW = total_BOW.sort_values(by=['Apariciones'], ascending=False)

total_tfidf = pd.DataFrame(TFIDF.mean(axis=0),columns=['Apariciones'])
total_tfidf = total_tfidf.sort_values(by=['Apariciones'], ascending=False)

display_side_by_side(total_BOW.head(20),total_tfidf.head(20), titles=['BOW','Tfidf'])
```

### Explicación

Para poder obtener los 20 tokens más frecuentes en las dos matrices he decido crear dos dataframes puesto que facilitaba a la hora de realizar el análisis. El motivo por el que no he utilizado el parámetro max_features es porque esté ordenaba los tokens por orden alfabético y no por frecuencia. El método para realizar la frecuencia en la matriz BOW ha sido mediante el acumulado total de apariciones de cada término. Por otro lado, para la matriz Tfidf esto se ha hecho con la media.

Para ambas matrices la mayoría de los términos más repetidos son los mismos. Se pueden observar algunas excepciones en el orden de frecuencia. Por ejemplo, para la matriz BOW el término "excelente" se encuentra en la posición 18 mientras que para la matriz Tfidf se encuentra en la 12. La palabra "desayuno" se encuentra entre el top 20 tokens más frecuentes para la matiz BOW mientras que para la matriz Tfidf no. La gran mayoría de estos términos son stop-words.

## 3.1 Ejercicio extra. Obtención de los 20 tokens más frecuentes sin tener en cuenta las stop-words

```python
stopwords = nlp.Defaults.stop_words
lista_hotel_stopwords=list(hotel["text"].apply(lambda x: ' '.join([word for word in x.split() if word not in (stopwords)])))

BOW_vectorizer_stopwords = CountVectorizer()
BOW_array_stopwords = BOW_vectorizer_stopwords.fit_transform(lista_hotel_stopwords).toarray()
vocab_BOW_stopwords = BOW_vectorizer_stopwords.get_feature_names()
BOW_20_stopwords=pd.DataFrame(BOW_array_stopwords, columns=vocab_BOW_stopwords)
BOW_20_stopwords
total_BOW_stopwords = pd.DataFrame(BOW_20_stopwords.sum(axis=0),columns=['Apariciones'])
total_BOW_stopwords = total_BOW_stopwords.sort_values(by=['Apariciones'], ascending=False)

Tfidf_vectorizer_stopwords = TfidfVectorizer()
Tfidf_array_stopwords = Tfidf_vectorizer_stopwords.fit_transform(lista_hotel_stopwords).toarray()
vocab_Tfidf_stopwords = Tfidf_vectorizer_stopwords.get_feature_names()
Tfidf_20_stopwords=pd.DataFrame(Tfidf_array_stopwords, columns=vocab_Tfidf_stopwords)
Tfidf_20_stopwords
total_Tfidf_stopwords = pd.DataFrame(Tfidf_20_stopwords.sum(axis=0),columns=['Apariciones'])
total_Tfidf_stopwords = total_Tfidf_stopwords.sort_values(by=['Apariciones'], ascending=False)

display_side_by_side(total_BOW_stopwords.head(20),total_Tfidf_stopwords.head(20), titles=['BOW','Tfidf']) 
```

### Explicación

He decidido añadir este ejercicio extra en el cual calculo los 20 tokens más frecuentes sin tener en cuenta los stop-words. Como se podía observar, estos eran los más frecuentes en el texto y no eran tan representativos a la hora de obtener una visión general de las opiniones de los clientes. EL problema que he podido detectar es que este código no es del todo óptimo puesto que se ve afectado por las faltas de ortografía de las reseñas. Es decir, hay varios tokens idénticos que se repiten en ambas listas puesto que no están escritos de la misma manera. Por ejemplo, "atención" o "habitación" se repiten porque ambos están escritos con tilde y sin tilde y el código los detecta como palabras distintas. Aún así, se pueden obtener unas ideas generales:

- Los clientes prestan una gran atención a la decoración, habitaciones, ubicación, personal o comida. Esto podría ayudar al director del hotel para centrarse en estos aspectos y mejorar la experiencia del cliente.
- La experiencia de los usuarios ha sido buena puesto que no aparecen términos negativos como "mal" o "peor" y si positivos como "excelente" o "agradable".
